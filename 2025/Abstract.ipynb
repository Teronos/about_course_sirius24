````python
def number_type_dict(number: float) -> str:
  converter = { 1: "positive", -1: "negative" }

  type_number = 0
  if number > 0:
    type_number = 1
  elif number < 0:
    type_number = -1

  str_repr = converter.get(type_number)

  if str_repr is None:
    return "zero"
  return str_repr

def number_type_dict_pretty(number: int|float) -> str| None:
  sign_extend = lambda x: 1 if x > 0 else (-1 if x < 0 else 0)

  converter = { 1: "positive", -1: "negative", 0: "zero" }

  return converter.get(sign_extend(number))
  #return converter[sign_extend(number)]


number_type_dict_pretty(-13)
````

````python
from typing import Callable, Any

conter2 = 0
def mul(a: int, b: int) -> int:
  global conter2
  conter2 += 1
  return a*b

counter1 = 0
def add(a: int, b: int) -> int:
  global counter1
  counter1 += 1
  return a + b

counter = 0
def Counter(func: Callable[..., Any]) -> Callable[..., Any]:
  def inner(*args: int, **kwargs: Any) -> Callable[..., Any] :
    global counter
    counter += 1
    return func(*args, **kwargs)
  return inner

function_counter = {}

def counter_smart(func: Callable[..., Any]) -> Callable[..., Any]:
  function_counter[func.__name__] = 0

  def gleb(*args: int, **kwargs: Any) -> Callable[..., Any]:
    global function_counter
    function_counter[func.__name__] += 1
    return func(*args, **kwargs)

  return gleb

@counter_smart
def add_simple(a: int, b: int) -> int:
  return a + b
#add_simple = Counter(add_simple)

@counter_smart
def multiply_func(a: int, b : int) -> int:
  return a * b

@counter_smart
def calc(action: Callable, *args) -> int:
  result = action(*args)
  print(f"result of calculation: {result}")
  return result


````

````python
import math
from typing import Self               # Python 3.11+
#from typing_extensions import Self     # Python 3.11<

class Complecs:
  __real: float
  __imag: float

  def __init__(self: Self, real: float, imag: float) -> None:
    self.__real = real
    self.__imag = imag

  def get_real(self: Self) -> float:
    return self.__real

  def get_imag(self) -> float:
    return self.__imag

  def __repr__(self) -> str:
    return f"{self.__real} + {self.get_imag()}i" # можно унифицировать

  def __add__(self, other_Complecs: Self) -> Self:
    real = 0
    imag = 0

    if isinstance(other_Complecs, Complecs):
      real = other_Complecs.get_real()
      imag = other_Complecs.get_imag()

    if isinstance(other_Complecs, (float, int)):
      real = other_Complecs

    return Complecs(
          self.__real + real,
          self.__imag + imag
          )

  def __eq__(self, other_Complecs: Self) -> bool:
    # if self.__real == other_Complecs.get_real() and self.__imag == other_Complecs.get_imag():
    #   return True
    # else:
    #   return False
    return (self.__real == other_Complecs.get_real()
      and self.__imag == other_Complecs.get_imag())

  def __len__(self) -> int:
    return int(round(math.dist((0, 0), (self.__real, self.__imag))))

  @staticmethod
  def cast_real(complecs) -> float:
    return complecs.get_real()

  @classmethod
  def form_dict(cls: Self, dct: dict) -> Self:
    real = dct.get("r", 0)
    imag = dct.get("i", 0)
    return cls(real, imag)

  @classmethod
  def from_list(cls, lst: list):
    if len(lst) != 2:
      raise Exception("len of list must be 2")
    return cls(lst[0], lst[1])
````

````python
class Vector(list):
  def __add__(self:Self, other: Self) -> Self:
    longest  = self

    if len(longest) < len(other):
      longest = other

    result = Vector([])
    for index, value in enumerate(longest):
      try:
        result.append(self[index] + other[index])
      except IndexError:
        result.append(value)

    return result

a = Vector([1, 2, 8, 99])
b = Vector([3, 4])

a + b
````

````python
def text_form_teplate(
    text: str, count_repit: int, sep:str, b_teg: str, e_teg: str
    ) -> str:

    text = reduce(
       lambda str1, str2: f"{str1}{str2}",
        [f"{text}{sep}" for _ in range(count_repit)]
       )
    return f"{b_teg}{text}{e_teg}"

single_repit = partial(text_form_teplate, count_repit = 1, sep = "")
triple_repit = partial(text_form_teplate, count_repit = 3, sep = "\n")

head_html_text = partial(single_repit, b_teg = "<H1>", e_teg = "</H1>")

print(head_html_text("Some text"))
print(triple_repit("Another text", b_teg = "<TAGG>", e_teg = "<ATCC>"))
````

````python
from typing import Any

class UnbrekableList:
  __elements: list = []
  def __init__(self, lst) -> None:
    self.__elements = lst

  def get_elem(self, index: int) -> Any:
    try:
      return self.__elements[index]
    except IndexError:
      return None

  def __repr__(self) -> str:
    return self.__elements.__repr__()

lst_custom = UnbrekableList([1, 2, 3])
print(lst_custom.get_elem(3))
print(lst_custom.get_elem(1))
````

обработчик событий

````python
class Event:
  __lifetime: float # seconds
  __state: int = 0 # seconds
  __request: str
  __ended: bool = False

  def __init__(self, lifetime: float, request: str) -> None:
    self.__lifetime = lifetime
    self.__request = request

  def execution(self) -> str:
    self.__state += 1
    if (self.__state >= self.__lifetime):
      self.__ended = True
      return f"{self.__request} is done"
    return f"{self.__request} working..."

  def __repr__(self) -> str:
    return f"{self.__request}"

  def is_done(self) -> bool:
    return self.__ended

class EventLoop:
  __events: list[Event] = []
  __state: int = 0

  def __init__(self, events: list[Event]) -> None:
    self.__events = events

  def __next__(self) -> Event:
    if len(self.__events) == 0:
      raise StopIteration
    event = self.__events[0]
    try:
      event = self.__events[self.__state]
      self.__state += 1
    except IndexError:
      self.__state = 0
      print("new cycle")
    finally:
      return event

  def __iter__(self):
    return self

  def __remove_event(self, index) -> None:
    try:
      self.__events.pop(index)
    except IndexError:
      pass

  def run(self) -> None:
    for event in self:
      print(event.execution())
      if event.is_done:
        print(f"{event} завершен")
        self.__remove_event(self.__state)


````

````python
event_list = [Event(i, f"{i}") for i in range(10)]
event_list
````

````python
loop = EventLoop(event_list)
loop.run()
````

````python
class City:
  name: str
  population: int
  exist_metro: bool
  airport: list[str]
  state_ecology: int
  young: int

  def __init__(self, name, population, exist_metro, airport, state_ecology, young):
    self.name = name
    self.population = population
    self.exist_metro = exist_metro
    self.airport = airport
    self.state_ecology = state_ecology
    self.young = young

  def rt_city(self):
    return ((self.conv_pop(self.population) + 10*self.exist_metro
     + 100*len(self.airport)) * 10*self.state_ecology * self.young)

  @staticmethod
  def conv_pop(population) -> int:
    mark = 0
    if population < 50_000:
      mark = 1
    elif population < 300_000:
      mark = 2
    elif population < 600_000:
      mark = 3
    elif population < 1_000_000:
      mark = 4
    else:
      mark = 5
    return mark

  def __repr__(self):
    return f"{self.name} {self.rt_city()}"

  def __lt__(self, other_City):
    return self.rt_city() < other_City.rt_city()
````

````python
Ms = City("Москва", 13_000_000, True, ["Внуково", "Шереметьево", "Домодедово",
                                       "Жуковский"], 4, 0.5)
Sm = City("Cмоленск", 330_000, False, [], 9, 0.35)
Vg = City("Воркута", 60_000, False, ["Воркута́"], 5, 0.3)
Hh = City("Шахты", 10_000, False, [], 10, 0.25)

Top_city = [Ms, Sm, Vg, Hh]
Top_city.sort()
Top_city
````
