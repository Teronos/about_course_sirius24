# Декораторы

````python
from typing import Callable, Any

conter2 = 0
def mul(a: int, b: int) -> int:
  global conter2
  conter2 += 1
  return a*b

counter1 = 0
def add(a: int, b: int) -> int:
  global counter1
  counter1 += 1
  return a + b

counter = 0
def Counter(func: Callable[..., Any]) -> Callable[..., Any]:
  def inner(*args: int, **kwargs: Any) -> Callable[..., Any] :
    global counter
    counter += 1
    return func(*args, **kwargs)
  return inner

function_counter = {}

def counter_smart(func: Callable[..., Any]) -> Callable[..., Any]:
  function_counter[func.__name__] = 0
  def inner(*args: int, **kwargs: Any) -> Callable[..., Any]:
    global function_counter
    function_counter[func.__name__] += 1
    return func(*args, **kwargs)
  return inner

@counter_smart
def add_simple(a: int, b: int) -> int:
  return a + b
#add_simple = Counter(add_simple)

@counter_smart
def multiply_func(a: int, b : int) -> int:
  return a * b
````

````python
function_counter = {}

def counter_with_initial(start: int) -> Callable[..., Any]:
  def wrapper(func: Callable[..., Any]) -> Callable[..., Any]:
      function_counter[func.__name__] = start
      def inner(*args: int, **kwargs: Any) -> Callable[..., Any]:
        global function_counter
        function_counter[func.__name__] += 1
        return func(*args, **kwargs)
      return inner
  return wrapper

@counter_with_initial(24)
def add_simple(a: int, b: int) -> int:
  return a + b
#add_simple = Counter(add_simple)

@counter_with_initial(3)
def multiply_func(a: int, b : int) -> int:
  return a * b
````

````python
import math
#from typing import Self               # Python 3.11+
from typing_extensions import Self     # Python 3.11<

class Complecs:
  __real: float
  __imag: float

  def __init__(self, real: float, imag: float) -> None:
    self.__real = real
    self.__imag = imag

  def get_real(self) -> float:
    return self.__real

  def get_imag(self) -> float:
    return self.__imag

  def __repr__(self) -> str:
    return f"{self.__real} + {self.get_imag()}i" # можно унифицировать

  def __add__(self, other_Complecs: Self) -> Self:
    real = 0
    imag = 0

    if isinstance(other_Complecs, Complecs):
      real = other_Complecs.get_real()
      imag = other_Complecs.get_imag()

    if isinstance(other_Complecs, (float, int)):
      real = other_Complecs

    return Complecs(
          self.__real + real,
          self.__imag + imag
          )

  def __eq__(self, other_Complecs: Self) -> bool:
    # if self.__real == other_Complecs.get_real() and self.__imag == other_Complecs.get_imag():
    #   return True
    # else:
    #   return False
    return (self.__real == other_Complecs.get_real()
      and self.__imag == other_Complecs.get_imag())

  def __len__(self) -> int:
    return int(round(math.dist((0, 0), (self.__real, self.__imag))))

  @staticmethod
  def cast_real(complecs) -> float:
    return complecs.get_real()

  @classmethod
  def form_dict(cls, dct: dict):
    real = dct.get("r", 0)
    imag = dct.get("i", 0)
    return cls(real, imag)

  @classmethod
  def from_list(cls, lst: list):
    if len(lst) != 2:
      raise Exception("len of list must be 2")
    return cls(lst[0], lst[1])
````

обработчик событий

````python
class Event:
  __lifetime: float # seconds
  __state: int = 0 # seconds
  __request: str
  __ended: bool = False

  def __init__(self, lifetime: float, request: str) -> None:
    self.__lifetime = lifetime
    self.__request = request

  def execution(self) -> str:
    self.__state += 1
    if (self.__state >= self.__lifetime):
      self.__ended = True
      return f"{self.__request} is done"
    return f"{self.__request} working..."

  def __repr__(self) -> str:
    return f"{self.__request}"

  def is_done(self) -> bool:
    return self.__ended

class EventLoop:
  __events: list[Event] = []
  __state: int = 0

  def __init__(self, events: list[Event]) -> None:
    self.__events = events

  def __next__(self) -> Event:
    if len(self.__events) == 0:
      raise StopIteration
    event = self.__events[0]
    try:
      event = self.__events[self.__state]
      self.__state += 1
    except IndexError:
      self.__state = 0
      print("new cycle")
    finally:
      return event

  def __iter__(self):
    return self

  def __remove_event(self, index) -> None:
    try:
      self.__events.pop(index)
    except IndexError:
      pass

  def run(self) -> None:
    for event in self:
      print(event.execution())
      if event.is_done:
        print(f"{event} завершен")
        self.__remove_event(self.__state)


````

````python
event_list = [Event(i, f"{i}") for i in range(10)]
event_list
````

````python
loop = EventLoop(event_list)
loop.run()
````

2) Упрошенный маркетплейс с пользователями, товарами, корзиной

````python
class City:
  name: str
  population: int
  exist_metro: bool
  airport: list[str]
  state_ecology: int
  young: int

  def __init__(self, name, population, exist_metro, airport, state_ecology, young):
    self.name = name
    self.population = population
    self.exist_metro = exist_metro
    self.airport = airport
    self.state_ecology = state_ecology
    self.young = young

  def rt_city(self):
    return ((self.conv_pop(self.population) + 10*self.exist_metro
     + 100*len(self.airport)) * 10*self.state_ecology * self.young)

  @staticmethod
  def conv_pop(population) -> int:
    mark = 0
    if population < 50_000:
      mark = 1
    elif population < 300_000:
      mark = 2
    elif population < 600_000:
      mark = 3
    elif population < 1_000_000:
      mark = 4
    else:
      mark = 5
    return mark

  def __repr__(self):
    return f"{self.name} {self.rt_city()}"

  def __lt__(self, other_City):
    return self.rt_city() < other_City.rt_city()



````

````python
Ms = City("Москва", 13_000_000, True, ["Внуково", "Шереметьево", "Домодедово",
                                       "Жуковский"], 4, 0.5)
Sm = City("Cмоленск", 330_000, False, [], 9, 0.35)
Vg = City("Воркута", 60_000, False, ["Воркута́"], 5, 0.3)
Hh = City("Шахты", 10_000, False, [], 10, 0.25)

Top_city = [Ms, Sm, Vg, Hh]
Top_city.sort()
Top_city
````
